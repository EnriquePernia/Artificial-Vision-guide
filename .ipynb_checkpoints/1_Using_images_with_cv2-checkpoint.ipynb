# 1. Using Images with cv2

```python
import numpy
import matplotlib.pyplot as plt
import cv2
```

## Import local images

Import local image using cv2 method `imread`.
* Import image and convert it to grayscale.
* Import directly grayscale image. 

**_good practice_**: use an absolute path to the image.
*Note*: Remember that `imgread` take the image using BGR as opposed to RGB. [see first plot]

```python
# color image
image_color =  cv2.imread('/home/betegon/Siali/jupyter_notebooks/paradrop.png') 
# convert to gray
image_converted_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)

# gray image
image_gray  =  cv2.imread('/home/betegon/Siali/jupyter_notebooks/paradrop.png', cv2.IMREAD_GRAYSCALE)
```

Represent images using `matplotlib`. It is possible to obtain pixels values and coordinates.

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 12))

# color image
axes[0].imshow(image_color)
axes[0].set_title('Color')

# gray image
axes[1].imshow(image_gray, cmap='gray')
axes[1].set_title('Gray')

axes[2].imshow(image_converted_gray, cmap='gray')
axes[2].set_title('Converted Gray')
```

As commented below, `color_image` is represented taking RGB instead of BGR. Time to change it:

Note the slightly different colors perceived in `image_gray` and `image_converted_gray`. The good one is the converted one, as it was converted in RGB.

```python
# Convert from BGR to RGB
image_color_RGB = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 3, figsize=(12, 12))

# color image
axes[0].imshow(image_color_RGB)
axes[0].set_title('Color')

# gray image of BGR, no bueno 
axes[1].imshow(image_gray, cmap='gray')
axes[1].set_title('Gray')

# converted gray image.
image_converted_gray = cv2.cvtColor(image_color_RGB, cv2.COLOR_BGR2GRAY)
axes[2].imshow(image_converted_gray, cmap='gray')
axes[2].set_title('Converted Gray')
```

## Write the images

In order to write the images is important to note that `cv2` method `imwrite` changes the input image to RGB as `cv2` thinks most of time it is better to write this way.

So in order to write the images correctly It is necessary to have the input image as BGR.

```python
# save images
image_color_BGR = cv2.cvtColor(image_color_RGB, cv2.COLOR_RGB2BGR)
# image_gray_BGR = cv2.cvtColor(image_gray, cv2.COLOR_RGB2BGR)

cv2.imwrite('image_color_BGR.png', image_color_BGR)
cv2.imwrite('image_gray.png', image_gray)
cv2.imwrite('image_converted_gray.png', image_converted_gray)
```

## Wrap up

Following points need to be clear in your mind to be able to advance in `cv2`:

* Importing images:
    *  Convert from BGR to RGB.
    
* Process images: 
    *  Do it with RGB image.
   
* Write images:
    *  Convert RGB to BGR in order to write it (`cv2` automatically converts BGR to RGB).
